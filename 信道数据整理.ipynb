```python
import pandas as pd 
import numpy as np
import os
np.set_printoptions(suppress=True)  #不用科学计数法表示
import warnings 
warnings.filterwarnings('ignore')
```

```python
def adjust_data(path):
    df = pd.read_csv(path)
    df = df.iloc[1::,:]
    df = df.reset_index(drop=True)
    freq = [];re_Trc1_S11 = [];im_Trc1_S11 = [];re_Trc2_S21 = [];im_Trc2_S21 = []; 
    for i in range(len(df) - 1):
        t = df.iloc[i + 1,0].split(';')
        freq.append(t[0])
        re_Trc1_S11.append(t[1])
        im_Trc1_S11.append(t[2])
        re_Trc2_S21.append(t[3])
        im_Trc2_S21.append(t[4])
    freq = pd.DataFrame(freq)
    re_Trc1_S11 = pd.DataFrame(re_Trc1_S11)
    im_Trc1_S11 = pd.DataFrame(im_Trc1_S11)
    re_Trc2_S21 = pd.DataFrame(re_Trc2_S21)
    im_Trc2_S21 = pd.DataFrame(im_Trc2_S21)
    data=pd.concat([freq, re_Trc1_S11, im_Trc1_S11, re_Trc2_S21, im_Trc2_S21], axis=1)
    data.columns=df.iloc[0,0].split(';')[0:5]
    return data

def fetch_data(rootdir):
    lst = os.listdir(rootdir)
    random_number = np.random.random_integers(0, len(lst), size = 1) #如想修改取的数据量，修改size的值就行
    print(random_number)
    random_files = [rootdir + '\\' + lst[i]  for i in random_number]
    random_data = []
    for i in range(len(random_files)):
        random_data += [random_files[i] + '\\' + j for j in os.listdir(random_files[i])]
    for i in range(len(random_data)):
        if i == 0:
            df = adjust_data(random_data[i])
        df=pd.concat([df, adjust_data(random_data[i])], axis=0)
    df.reset_index(drop=True)
    return df.reset_index(drop=True)
```
